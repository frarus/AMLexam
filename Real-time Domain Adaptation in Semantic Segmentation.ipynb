```python
# GET INFO ABOUT LABELS AND THEIR COLORS TO PRINT A LEGEND
# info.json must be on Google Drive in the folder Datasets/Cityscapes
import json
import numpy as np
import matplotlib.pyplot as plt

infojson = json.load(open('/content/drive/MyDrive/Datasets/Cityscapes/info.json', 'r'))
colors=infojson['palette']
classes=infojson['label']
f=lambda m,c: plt.plot([],[], marker=m, color=c, ls="none")[0]
colors=np.array(np.array(colors)/255, dtype=np.float32)
handles=[f("s", colors[i]) for i in range(20)]
legend=plt.legend(handles, classes, loc=3, frameon=False)
fig = legend.figure
fig.canvas.draw()
bbox=legend.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
plt.gca().set_axis_off()
plt.show()
```
